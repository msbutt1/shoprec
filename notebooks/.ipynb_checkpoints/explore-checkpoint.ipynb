# ShopRec: Data Exploration and Visualization

This notebook explores the ShopRec recommendation system by:
1. Loading and examining the fake purchase dataset
2. Visualizing the user-product interaction matrix
3. Analyzing SVD components
4. Testing live inference for multiple users

Let's start by importing the necessary libraries and setting up the environment.


```python
import sys
from pathlib import Path
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.sparse import csr_matrix
import joblib

# Add project root to Python path
project_root = Path().resolve().parent
sys.path.insert(0, str(project_root))

# Import ShopRec modules
from src.recommender.utils import load_csv_to_matrix, load_model_artifacts
from src.recommender.hybrid import create_hybrid_recommender
from src.recommender.infer import recommend_products_for_user

# Set up plotting style
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")
%matplotlib inline

print("✓ Imports successful!")
print(f"Project root: {project_root}")
```

## 1. Load Fake Purchase Dataset

First, let's load the fake purchase data and examine its structure. This dataset contains user-product interactions that we'll use to train and evaluate our recommendation model.


```python
# Load the fake purchase dataset
csv_path = project_root / "data" / "fake_purchases.csv"
df = pd.read_csv(csv_path)

print(f"Dataset shape: {df.shape}")
print(f"\nFirst few rows:")
print(df.head(10))
print(f"\nDataset info:")
print(df.info())
print(f"\nBasic statistics:")
print(df.describe())
```

```python
# Analyze the dataset
print("=== Dataset Analysis ===\n")
print(f"Total purchases: {len(df)}")
print(f"Unique users: {df['user_id'].nunique()}")
print(f"Unique products: {df['product_id'].nunique()}")
print(f"Date range: {df['timestamp'].min()} to {df['timestamp'].max()}")

# Check for duplicates
duplicates = df.duplicated(subset=['user_id', 'product_id']).sum()
print(f"Duplicate user-product pairs: {duplicates}")

# User activity distribution
user_activity = df.groupby('user_id').size()
print(f"\nUser activity stats:")
print(f"  Min purchases per user: {user_activity.min()}")
print(f"  Max purchases per user: {user_activity.max()}")
print(f"  Mean purchases per user: {user_activity.mean():.2f}")

# Product popularity
product_popularity = df.groupby('product_id').size()
print(f"\nProduct popularity stats:")
print(f"  Min purchases per product: {product_popularity.min()}")
print(f"  Max purchases per product: {product_popularity.max()}")
print(f"  Mean purchases per product: {product_popularity.mean():.2f}")
```

## 2. Build User-Product Interaction Matrix

The interaction matrix is the core data structure for collaborative filtering. Each row represents a user, each column represents a product, and the values indicate whether (or how many times) a user has purchased a product.


```python
# Build the user-product interaction matrix
matrix, user_id_to_idx, product_id_to_idx = load_csv_to_matrix(
    csv_path=str(csv_path),
    user_col="user_id",
    item_col="product_id",
    binary=True,
)

print(f"Matrix shape: {matrix.shape} (users × products)")
print(f"Matrix density: {(matrix.nnz / (matrix.shape[0] * matrix.shape[1]) * 100):.2f}%")
print(f"Non-zero entries: {matrix.nnz}")
print(f"\nUser ID to index mapping (first 5):")
for i, (uid, idx) in enumerate(list(user_id_to_idx.items())[:5]):
    print(f"  User {uid} → Index {idx}")
print(f"\nProduct ID to index mapping (first 5):")
for i, (pid, idx) in enumerate(list(product_id_to_idx.items())[:5]):
    print(f"  Product {pid} → Index {idx}")
```

## 3. Visualize User-Product Matrix Heatmap

A heatmap helps us understand the sparsity and patterns in user-product interactions. Since the matrix is sparse, we'll visualize a subset of users and products to make the patterns visible.


```python
# Convert sparse matrix to dense for visualization (subset)
# Select a subset of users and products for better visualization
n_users_viz = min(30, matrix.shape[0])
n_products_viz = min(50, matrix.shape[1])

matrix_dense = matrix[:n_users_viz, :n_products_viz].toarray()

# Create heatmap
fig, ax = plt.subplots(figsize=(14, 8))
sns.heatmap(
    matrix_dense,
    cmap='YlOrRd',
    cbar_kws={'label': 'Purchase (1=Yes, 0=No)'},
    xticklabels=False,
    yticklabels=False,
    ax=ax
)
ax.set_title(f'User-Product Interaction Matrix Heatmap\n(First {n_users_viz} users × First {n_products_viz} products)', 
             fontsize=14, fontweight='bold')
ax.set_xlabel('Products', fontsize=12)
ax.set_ylabel('Users', fontsize=12)
plt.tight_layout()
plt.show()

print(f"Visualized {n_users_viz} users × {n_products_viz} products")
print(f"White areas = No interaction, Colored areas = Purchase occurred")
```

```python
# Additional visualization: User activity and product popularity
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# User activity distribution
user_purchase_counts = np.array(matrix.sum(axis=1)).flatten()
axes[0].hist(user_purchase_counts, bins=20, edgecolor='black', alpha=0.7)
axes[0].set_xlabel('Number of Purchases per User', fontsize=11)
axes[0].set_ylabel('Number of Users', fontsize=11)
axes[0].set_title('User Activity Distribution', fontsize=12, fontweight='bold')
axes[0].grid(True, alpha=0.3)

# Product popularity distribution
product_purchase_counts = np.array(matrix.sum(axis=0)).flatten()
axes[1].hist(product_purchase_counts, bins=20, edgecolor='black', alpha=0.7, color='orange')
axes[1].set_xlabel('Number of Purchases per Product', fontsize=11)
axes[1].set_ylabel('Number of Products', fontsize=11)
axes[1].set_title('Product Popularity Distribution', fontsize=12, fontweight='bold')
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print(f"User activity: min={user_purchase_counts.min()}, max={user_purchase_counts.max()}, mean={user_purchase_counts.mean():.2f}")
print(f"Product popularity: min={product_purchase_counts.min()}, max={product_purchase_counts.max()}, mean={product_purchase_counts.mean():.2f}")
```

## 4. Load Trained Model and Analyze SVD Components

The SVD (Singular Value Decomposition) model decomposes the user-product matrix into latent features. These components capture underlying patterns in user preferences and product characteristics. Let's examine what the model has learned.


```python
# Load the trained model
model_dir = project_root / "models" / "hybrid"  # Using hybrid model with embeddings
model, user_map, product_map = load_model_artifacts(str(model_dir))

print(f"Model loaded successfully!")
print(f"Model type: {type(model).__name__}")
print(f"Number of components: {model.n_components}")
print(f"Components shape: {model.components_.shape}")
print(f"Explained variance ratio: {model.explained_variance_ratio_.sum():.4f}")
print(f"\nFirst few explained variance ratios:")
for i, var in enumerate(model.explained_variance_ratio_[:10]):
    print(f"  Component {i}: {var:.4f}")
```

### 4.1 Visualize Explained Variance

The explained variance ratio tells us how much of the original data variance each component captures. Components with higher explained variance are more important.


```python
# Plot explained variance
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Cumulative explained variance
cumulative_var = np.cumsum(model.explained_variance_ratio_)
axes[0].plot(range(1, len(cumulative_var) + 1), cumulative_var, marker='o', linewidth=2, markersize=4)
axes[0].axhline(y=0.8, color='r', linestyle='--', label='80% variance')
axes[0].axhline(y=0.9, color='g', linestyle='--', label='90% variance')
axes[0].set_xlabel('Number of Components', fontsize=11)
axes[0].set_ylabel('Cumulative Explained Variance', fontsize=11)
axes[0].set_title('Cumulative Explained Variance by Component', fontsize=12, fontweight='bold')
axes[0].legend()
axes[0].grid(True, alpha=0.3)

# Individual explained variance (first 20 components)
n_show = min(20, len(model.explained_variance_ratio_))
axes[1].bar(range(1, n_show + 1), model.explained_variance_ratio_[:n_show], alpha=0.7)
axes[1].set_xlabel('Component Number', fontsize=11)
axes[1].set_ylabel('Explained Variance Ratio', fontsize=11)
axes[1].set_title(f'Individual Explained Variance (First {n_show} Components)', fontsize=12, fontweight='bold')
axes[1].grid(True, alpha=0.3, axis='y')

plt.tight_layout()
plt.show()

print(f"Total variance explained: {cumulative_var[-1]:.4f}")
print(f"Components needed for 80% variance: {np.argmax(cumulative_var >= 0.8) + 1}")
print(f"Components needed for 90% variance: {np.argmax(cumulative_var >= 0.9) + 1}")
```

### 4.2 Visualize SVD Components

The SVD components represent latent features that capture patterns in the data. Each component is a vector across all products, showing which products are similar in the latent space.


```python
# Visualize the first few SVD components
n_components_show = min(6, model.n_components)
n_products_show = min(30, model.components_.shape[1])

fig, axes = plt.subplots(2, 3, figsize=(16, 10))
axes = axes.flatten()

for i in range(n_components_show):
    component = model.components_[i, :n_products_show]
    axes[i].bar(range(n_products_show), component, alpha=0.7)
    axes[i].set_title(f'Component {i+1}\n(Variance: {model.explained_variance_ratio_[i]:.4f})', 
                     fontsize=11, fontweight='bold')
    axes[i].set_xlabel('Product Index', fontsize=10)
    axes[i].set_ylabel('Component Value', fontsize=10)
    axes[i].grid(True, alpha=0.3, axis='y')
    axes[i].axhline(y=0, color='black', linestyle='-', linewidth=0.5)

plt.suptitle('First 6 SVD Components (Product Latent Features)', fontsize=14, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()

print("Each component represents a latent feature that captures patterns across products.")
print("Positive values indicate products that are similar in this latent dimension.")
```

```python
# Heatmap of all components (subset of products)
n_products_heatmap = min(40, model.components_.shape[1])
components_heatmap = model.components_[:, :n_products_heatmap]

fig, ax = plt.subplots(figsize=(14, 8))
sns.heatmap(
    components_heatmap,
    cmap='RdBu_r',
    center=0,
    cbar_kws={'label': 'Component Value'},
    xticklabels=False,
    yticklabels=[f'Comp {i+1}' for i in range(model.n_components)],
    ax=ax
)
ax.set_title(f'SVD Components Heatmap\n({model.n_components} components × {n_products_heatmap} products)', 
             fontsize=14, fontweight='bold')
ax.set_xlabel('Products', fontsize=12)
ax.set_ylabel('Components', fontsize=12)
plt.tight_layout()
plt.show()

print("This heatmap shows how each component (row) relates to each product (column).")
print("Red = positive values, Blue = negative values. Similar colors indicate similar patterns.")
```

## 5. Live Inference: Test Recommendations for Multiple Users

Now let's test the recommendation system in action! We'll generate recommendations for several users and analyze what the model suggests.


```python
# Create hybrid recommender
hybrid_recommender = create_hybrid_recommender(
    model=model,
    user_id_to_idx=user_map,
    product_id_to_idx=product_map,
    model_dir=str(model_dir),
    cf_weight=0.7,
    content_weight=0.3,
)

print("✓ Hybrid recommender created!")
print(f"  CF weight: 0.7 (70%)")
print(f"  Content weight: 0.3 (30%)")
```

```python
# Select 4 users for testing
test_users = list(user_map.keys())[:4]  # First 4 users in the model
top_n = 10

print("=== Testing Recommendations for 4 Users ===\n")
print("Note: Passing user_product_matrix to enable content-based scoring\n")

for user_id in test_users:
    # Get recommendations with explainability
    # Pass the matrix so content-based scoring can use purchase history
    recommendations, scores = hybrid_recommender.recommend(
        user_id=user_id,
        top_n=top_n,
        user_product_matrix=matrix,  # Enable content-based scoring with purchase history
        return_scores=True,
    )
    
    print(f"\n{'='*60}")
    print(f"User {user_id} Recommendations (Top {len(recommendations)}):")
    print(f"{'='*60}")
    print(f"Recommended Products: {recommendations}")
    
    # Show score breakdown for top 5
    if scores and 'hybrid_scores' in scores:
        print(f"\nScore Breakdown (Top 5):")
        top_5_recs = recommendations[:5]
        for pid in top_5_recs:
            cf_score = scores['cf_scores'].get(pid, 0.0)
            content_score = scores['content_scores'].get(pid, 0.0)
            hybrid_score = scores['hybrid_scores'].get(pid, 0.0)
            print(f"  Product {pid}: CF={cf_score:.3f}, Content={content_score:.3f}, Hybrid={hybrid_score:.3f}")
    
    # Check user's purchase history
    if user_id in user_id_to_idx:
        user_idx = user_id_to_idx[user_id]
        user_purchases = matrix[user_idx].nonzero()[1]
        purchased_product_ids = [list(product_id_to_idx.keys())[idx] for idx in user_purchases]
        print(f"\nUser's Purchase History ({len(purchased_product_ids)} products):")
        print(f"  {purchased_product_ids[:10]}{'...' if len(purchased_product_ids) > 10 else ''}")
```

### 5.1 Visualize Recommendations

Let's create a visualization comparing the recommendations for different users to see how personalized they are.


```python
# Collect recommendations for all test users
all_recommendations = {}
for user_id in test_users:
    recommendations = hybrid_recommender.recommend(
        user_id=user_id,
        top_n=top_n,
        user_product_matrix=matrix,  # Enable content-based scoring
        return_scores=False,
    )
    all_recommendations[user_id] = set(recommendations)

# Create visualization
fig, ax = plt.subplots(figsize=(12, 6))

# Calculate overlap between users
n_users = len(test_users)
overlap_matrix = np.zeros((n_users, n_users))

for i, user1 in enumerate(test_users):
    for j, user2 in enumerate(test_users):
        if i == j:
            overlap_matrix[i, j] = 1.0  # Same user
        else:
            rec1 = all_recommendations[user1]
            rec2 = all_recommendations[user2]
            if len(rec1) > 0 and len(rec2) > 0:
                overlap = len(rec1 & rec2) / len(rec1 | rec2)  # Jaccard similarity
                overlap_matrix[i, j] = overlap

# Plot heatmap
sns.heatmap(
    overlap_matrix,
    annot=True,
    fmt='.2f',
    cmap='YlOrRd',
    xticklabels=[f'User {uid}' for uid in test_users],
    yticklabels=[f'User {uid}' for uid in test_users],
    cbar_kws={'label': 'Recommendation Overlap (Jaccard)'},
    ax=ax
)
ax.set_title('Recommendation Overlap Between Users\n(Higher values = more similar recommendations)', 
             fontsize=12, fontweight='bold')
plt.tight_layout()
plt.show()

print("Jaccard similarity = (Intersection) / (Union)")
print("1.0 = identical recommendations, 0.0 = completely different recommendations")
```

```python
# Compare CF-only vs Hybrid recommendations
user_id = test_users[0]  # Use first user

# CF-only recommendations
from src.recommender.infer import _compute_recommendations
user_idx = user_map[user_id]
cf_recommendations = _compute_recommendations(
    user_idx=user_idx,
    model=model,
    product_id_to_idx=product_map,
    top_n=top_n,
    user_product_matrix=matrix,  # Use actual purchase history for fair comparison
)

# Hybrid recommendations
hybrid_recommendations = hybrid_recommender.recommend(
    user_id=user_id,
    top_n=top_n,
    user_product_matrix=matrix,  # Enable content-based scoring
    return_scores=False,
)

print(f"=== Comparison for User {user_id} ===\n")
print(f"CF-only recommendations ({len(cf_recommendations)}):")
print(f"  {cf_recommendations}\n")
print(f"Hybrid recommendations ({len(hybrid_recommendations)}):")
print(f"  {hybrid_recommendations}\n")

# Calculate overlap
cf_set = set(cf_recommendations)
hybrid_set = set(hybrid_recommendations)
overlap = len(cf_set & hybrid_set)
union = len(cf_set | hybrid_set)
jaccard = overlap / union if union > 0 else 0.0

print(f"Overlap: {overlap} products")
print(f"Jaccard similarity: {jaccard:.3f}")
print(f"\nProducts only in CF: {list(cf_set - hybrid_set)}")
print(f"Products only in Hybrid: {list(hybrid_set - cf_set)}")
```

### 5.2 Test Cold-Start Scenario

Let's test what happens when we try to get recommendations for a user who wasn't in the training data (cold-start problem).


```python
# Test with a new user (not in training data)
cold_start_user_id = 99999  # A user ID that doesn't exist

print(f"=== Cold-Start Test: User {cold_start_user_id} ===\n")
print(f"User in training data: {cold_start_user_id in user_map}")

# Get recommendations
cold_start_recommendations = hybrid_recommender.recommend(
    user_id=cold_start_user_id,
    top_n=top_n,
    return_scores=False,
)

print(f"\nCold-start recommendations ({len(cold_start_recommendations)}):")
print(f"  {cold_start_recommendations}")
print(f"\nNote: Cold-start users receive default/popular product recommendations")
print(f"      since we don't have their purchase history.")
```
